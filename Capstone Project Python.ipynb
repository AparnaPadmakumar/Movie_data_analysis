```python
#treating missing value by replacing it by mode
def treating_missing_value(dataframe):
    for column in dataframe.columns:
        mode=dataframe[column].mode()[0]
        dataframe[column].fillna(mode,inplace=True)
treating_missing_value(df)
```

## TASK 10

```python
#Sorting the data in ascending order based on revenue
sorted_data =df.sort_values('revenue', ascending=False)
#Filter the top 500 movies based on revenue
top_500_movies =sorted_data.head(500)
```

```python
#central tendency for budget
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().values[0]
print('Budget:')
print("Mean: ",budget_mean)
print("Median: ",budget_median)
print("Mode: ", budget_mode)
```

```python
#central tendency for revenue
revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().values[0]
print('Revenue:')
print("Mean: ",revenue_mean)
print("Median: ",revenue_median)
print("Mode: ", revenue_mode)
```

```python
#central tendency for runtime
runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().values[0]
print('Runtime:')
print("Mean: ",runtime_mean)
print("Median: ",runtime_median)
print("Mode: ", runtime_mode)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([top_500_movies['budget'], top_500_movies['revenue'], top_500_movies['runtime']])
plt.xticks([1, 2, 3], ['Budget', 'Revenue', 'Runtime'])
plt.ylabel('Value')
plt.title('Outlier Analysis')
plt.show()
```

```python
#Finding and removing outlier from Budget and Runtime
outlier_column=['budget', 'revenue', 'runtime']
for column in outlier_column:
    Q1=top_500_movies[column].quantile(0.25)
    Q3=top_500_movies[column].quantile(0.75)
    IQR=Q3-Q1
    LF=Q1-1.5*IQR
    UF=Q3+1.5*IQR
    top_500_movies=top_500_movies[(top_500_movies[column]<UF) & (top_500_movies[column]>LF)]
```

## TASK 11

```python
#movies above average run time
above_average_runtime_movies =top_500_movies[top_500_movies['runtime'] > runtime_mean]
above_average_runtime_movies[['original_title','runtime']]
```
